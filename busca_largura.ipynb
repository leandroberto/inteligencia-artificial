```python
from copy import deepcopy
import math
```

```python
class Estado:
    def __init__(self,vetor = [], custo = 0, acao = None, pai = None):
        self.vetor = vetor
        self.custo = custo
        self.acao = acao
        self.pai = pai
    
    def get_custo(self):
        return self.custo
```

```python
def movimentos(estado):
    listax=deepcopy(estado.vetor)
    last_ta=0
    last_tb=0
    last_tc=0
    
    custo_AB=1
    custo_AC=1
    custo_BA=1
    custo_BC=1
    custo_CA=1
    custo_CB=1
    
    #Ultimo elemento da lista Torre A
    if(len(listax[0])==0):
        last_ta=math.inf
    else:
        last_ta=listax[0][-1]        
    #Ultimo elemento da lista Torre B
    if(len(listax[1])==0):
        last_tb=math.inf
    else:
        last_tb=listax[1][-1]        
    #Ultimo elemento da lista Torre C
    if(len(listax[2])==0):
        last_tc=math.inf
    else:
        last_tc=listax[2][-1]
    
    #Movimento A->B
    mov_AB=deepcopy(listax) 
    if(last_ta<last_tb and last_ta!=math.inf):       
        mov_AB[1].append(mov_AB[0].pop())
    else:
        mov_AB=[]
    #Movimento A->C
    mov_AC=deepcopy(listax) 
    if(last_ta<last_tc and last_ta!=math.inf):       
        mov_AC[2].append(mov_AC[0].pop())
    else:
        mov_AC=[]
    #Movimento B->A
    mov_BA=deepcopy(listax) 
    if(last_tb<last_ta and last_tb!=math.inf):       
        mov_BA[0].append(mov_BA[1].pop())
    else:
        mov_BA=[]
    #Movimento B->C
    mov_BC=deepcopy(listax) 
    if(last_tb<last_tc and last_tb!=math.inf):       
        mov_BC[2].append(mov_BC[1].pop())
    else:
        mov_BC=[]
    #Movimento C->A
    mov_CA=deepcopy(listax) 
    if(last_tc<last_ta and last_tc!=math.inf):       
        mov_CA[0].append(mov_CA[2].pop())
    else:
        mov_CA=[]
    #Movimento C->B
    mov_CB=deepcopy(listax)
    if(last_tc<last_tb and last_tc!=math.inf):       
        mov_CB[1].append(mov_CB[2].pop())
    else:
        mov_CB=[]

    filhos=[]
    if mov_AB!=[]:
        filho_AB=Estado(mov_AB,custo_AB,"A->B",listax)
        filhos.append(filho_AB)
    if mov_AC!=[]:
        filho_AC=Estado(mov_AC,custo_AC,"A->C",listax)
        filhos.append(filho_AC)
    if mov_BA!=[]:
        filho_BA=Estado(mov_BA,custo_BA,"B->A",listax)
        filhos.append(filho_BA)
    if mov_BC!=[]:
        filho_BC=Estado(mov_BC,custo_BC,"B->C",listax)
        filhos.append(filho_BC)
    if mov_CA!=[]:
        filho_CA=Estado(mov_CA,custo_CA,"C->A",listax)
        filhos.append(filho_CA)
    if mov_CB!=[]:
        filho_CB=Estado(mov_CB,custo_CB,"C->B",listax)
        filhos.append(filho_CB)
    return(filhos)
        
```

```python
def print_lista(lista):
    for i in range(len(lista)):
        print(lista[i].custo,end="")
        print(",",end="")
    print("")
```

```python
def imprimir_solucao(estado_inicial,solucao,conjunto_solucao,historico,tipo_busca):
    estado_atual=deepcopy(conjunto_solucao)
    arvore=[]
    arvore.append(estado_atual)
    while(estado_atual.pai!=None):
        for i in range(len(historico)):
            if(historico[i].vetor==estado_atual.pai or estado_atual.pai==None):
                arvore.insert(0,historico[i])
                estado_atual=historico[i]
                break
    print("")
    print("Tipo de Busca: Em "+tipo_busca)
    print("Movimentos: ",end="")
    print(len(arvore)-1)
    print("----------AÇÕES----------")
    for i in range(len(arvore)):
        if(arvore[i].acao!=None):
            print(arvore[i].acao,end="")
            if(i!=len(arvore)-1):
                print(" ",end="")
    print("")
    print("----------CONJUNTOS----------")
    for i in range(len(arvore)):
        print(arvore[i].vetor,end="")
        if(i!=len(arvore)-1):
            print("  ->  ",end="")
    print("")
```

```python
def busca_largura(estado,solucao):
    estado_inicial=deepcopy(estado)
    historico=[];#conjutos já vistos
    fronteira=[]
    n_iteracoes=0
    novo_conjunto=True
    solucao_encontrada=False
    conjunto_solucao=0
    while(solucao_encontrada==False):       
        historico.append(estado)
        filhos=movimentos(estado)        
        for i in range(len(filhos)):
            for j in range(len(historico)):
                if(filhos[i].vetor==historico[j].vetor):
                    novo_conjunto=False
                    break
            if(novo_conjunto==True):
                if(filhos[i].vetor==solucao.vetor):
                    solucao_encontrada=True
                    conjunto_solucao=deepcopy(filhos[i])
                    break                    
                fronteira.append(filhos[i])
                historico.append(filhos[i])
            novo_conjunto=True
        if(solucao_encontrada!=True):
            estado=fronteira.pop(0)
    imprimir_solucao(estado_inicial,solucao,conjunto_solucao,historico,"Largura")
```

```python
qtd_discos=int(input("Informe a quantidade de Discos: "))
torres=[[],[],[]]
v_solucao=[[],[],[]]
for i in range(qtd_discos,0,-1):
    torres[0].append(i)
    v_solucao[2].append(i)
estado_inicial=Estado(torres,0,None,None)
solucao=Estado(v_solucao)
print("Estado Inicial:",end="")
print(estado_inicial.vetor)
print("Solucao Desejada:",end="")
print(solucao.vetor)
busca_largura(estado_inicial,solucao)
```
